2.a program which will find all such numbers which are divisible by 7 but are not a
multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed
in a comma-separated sequence on a single line.

```python
def IsDivisableBy7(number):
    if (number%7 == 0):
        return True 
```

```python
def IsMultipleOf5(number):
    if(number%5 == 0):
        return True 
```

```python
#using while
finalstring = ''
startNum = 2000
while (startNum <= 3200): 
    if(IsDivisableBy7(startNum) and not IsMultipleOf5(startNum)):
        finalstring = finalstring+str(startNum) +','        
    startNum = startNum +1
    
print(finalstring.strip(','))
```

```python
#Usinng for loop
finalvalue =''
for num in range(2000,3201):
    if(IsDivisableBy7(num) and not IsMultipleOf5(num)):
        finalvalue =  finalvalue + str(num)+ ','
print(finalvalue.strip(','))
```

```python
#using list comprehention
lst = [x for x in range(2000,3201) if(IsDivisableBy7(x) and not IsMultipleOf5(x))]
str(lst)
```

3.Write a Python program to accept the user's first and last name and then getting them
printed in the the reverse order with a space between first name and last name.

```python
FirstName = input("Enter First name: ")
LastName = input("Enter Last name: ")

print(FirstName[::-1] + " " + LastName[::-1])
```

4.Write a Python program to find the volume of a sphere with diameter 12 cm.
Formula: V=4/3 *  * r 3

```python
def CalculateValumeOfSphere(diameter):
    radius = diameter/2
    volume = 4/3 * 3.14 * pow(radius,3)
    return volume

print(CalculateValumeOfSphere(8))
```
